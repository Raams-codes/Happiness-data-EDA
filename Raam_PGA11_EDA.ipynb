```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(data=df,x="Ladder score",color="g",stat="percent").set(title="Overall Happiness Percentage ")
```

```python
sns.histplot(data=df,x="Ladder score",stat="percent",hue="Regional indicator", multiple="stack",element="bars")
```

```python
# Insights:
    most of sub saharan and south asian region is less in happiness score (less than 5)
    All the western europion countries are high in happiness score(above 6)
    north american countreis fall in average happiness score(5-7)
    most of east asian countries are in average happiness score (around 5) 
    commenwealth independent states are in 5 to 6
    middle east and african contries are falling in wide range of disparity among countries (3 to 7)
```

```python
plt.figure(figsize=(12,120))
sns.barplot(data=df,x="Ladder score",y=df["Country name"],hue="Regional indicator")
```

```python
sns.pairplot(df[['Ladder score',"Regional indicator",'Logged GDP per capita', 'Social support', 'Healthy life expectancy','Freedom to make life choices', 'Generosity','Perceptions of corruption']],hue='Regional indicator')
```

```python
plt.figure(figsize=(25,25))
sns.pairplot(df,x_vars=['Logged GDP per capita', 'Social support', 'Healthy life expectancy','Freedom to make life choices',
                        'Generosity','Perceptions of corruption'],y_vars='Ladder score',hue='Regional indicator')
```

```python
#Insights
     Most of western europian countries are having high GDP as well as happiness score
     Most of sub-saharan region having less GDP and Happiness score
     Latin-American and caribbean countries are average score in freedom to make choices,healthy life expectancty as well as happiness score
     south asian countries which have good social support and GDP are having good happiness score and countries with less healthy life expectancy  has less ladder scor 
```

```python
correlation=df[['Ladder score','Logged GDP per capita', 'Social support', 'Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']].corr()
correlation
```

```python
df_corr=pd.DataFrame({"c_columns":correlation.columns,"corr":correlation['Ladder score']})
df_corr
```

```python
p=sns.barplot(data=df_corr,y="c_columns",x="corr")
p.set(title="Correlation with happiness score")
```

```python
#Insights:
    ladder score directly propotional to GDP, Social support,Healthy life expectancy
    freedom to make chances is little less propotional to happiness score 
    generosity have no correlation between happiness score
    perceptions of corruption considerablly impacts happiness score
```

```python
sns.heatmap(correlation,annot=True)
```
